```python
# Import necessary libraries
import pandas as pd
import numpy as np
import joblib
import shap
import lime
import lime.lime_tabular
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load the trained logistic regression model
model_path = '/content/drive/My Drive/IDS models/NF-ToN-IoT-V2_logistic_regression_model.pkl'
logreg = joblib.load(model_path)

# Load your dataset for explainability (use the test dataset)
test_data = pd.read_csv('/content/drive/My Drive/IDS Dataset 2/NF-ToN-IoT-V2_15000 samples_minmax_normalized_test.csv')

# Separate features and labels
X_test = test_data.iloc[:, :-2]  # All columns except the last two (labels)
y_test = test_data.iloc[:, -1]    # Last column (label names)

# SHAP Explainability
# Initialize SHAP explainer using KernelExplainer
explainer = shap.KernelExplainer(logreg.predict_proba, X_test)

# Calculate SHAP values for the test set (optional: specify a subset for faster computation)
shap_values = explainer.shap_values(X_test[:10])  # Compute for first 100 samples

# Plot SHAP summary plot
#shap.summary_plot(shap_values, X_test[:10])  # Plot for first 100 samples
```

```python
class_names = y_test.unique()  # Get unique values from the target variable
class_names
```

```python
import matplotlib.pyplot as plt

# Plotting with adjusted figure size
num_classes = len(class_names)  # Get the correct number of classes

for class_index in range(num_classes):
    shap_values_class = shap_values[:, :, class_index]

    # Set figure size before plotting
    plt.figure(figsize=(10, 6))  # Adjust width and height as needed

    # Use the same subset of X_test that was used for SHAP calculation
    X_test_subset = X_test.iloc[:10]

    shap.summary_plot(shap_values_class, X_test_subset, feature_names=X_test_subset.columns,
                      max_display=10, show=False)
    class_name = class_names[class_index]
    plt.title(f"SHAP Summary Plot for Class: {class_name}")
```

```python
# LIME Explainability
# Initialize LIME explainer
lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_test),
    feature_names=X_test.columns,
    class_names=np.unique(y_test),
    mode='classification'
)

# Choose an instance to explain (for example, the first instance)
instance_to_explain = X_test.iloc[698]

# Generate LIME explanation
lime_exp = lime_explainer.explain_instance(
    data_row=instance_to_explain,
    predict_fn=logreg.predict_proba
)

# Plot LIME explanation for the selected instance
lime_exp.as_pyplot_figure()
plt.title('LIME Explanation for Instance 698')
plt.show()
```
